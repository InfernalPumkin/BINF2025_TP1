Q9. Combien de groupes sont présents par royaume ?

````python
def group_kingdom(file, kingdom):
    group = []
    with open(file, "r") as f:
        f.readline()
        for count, lines in enumerate(f):
            if kingdom in lines:
                info = lines.split("\t")
                if len(info) >= 3:
                    if not info[2] in group:
                        group.append(info[2])
    return group                
    
Bacteria_group = group_kingdom("overview.txt", "Bacteria")
Archaea_group = group_kingdom("overview.txt", "Archaea")
Eukaryota_group = group_kingdom("overview.txt", "Eukaryota")
Viruses_group = group_kingdom("overview.txt", "Viruses")
print('Bacteria:', len(Bacteria_group))
print('Archaea:', len(Archaea_group))
print(Archaea_group)
print('Eukaryota:', len(Eukaryota_group))
print('Viruses:', len(Viruses_group))
````
